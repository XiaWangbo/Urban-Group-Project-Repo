```python
import os
import pandas as pd

# 设置文件夹路径
folder_path = 'hs300_market_data'

# 获取文件夹中的所有CSV文件
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# 初始化一个空的DataFrame来存储数据
merged_data = pd.DataFrame()

# 读取并合并所有CSV文件
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    temp_data = pd.read_csv(file_path)
    merged_data = pd.concat([merged_data, temp_data], ignore_index=True)

# 保存合并后的数据为一个新的CSV文件
merged_data.to_csv('hs300_merged_market_data.csv', index=False)
print('Saved merged market data to hs300_merged_market_data.csv')
```

```python
import pandas as pd
import numpy as np

# 读取数据
data = pd.read_csv('hs300_preprocessed_market_data.csv')
data['date'] = pd.to_datetime(data['date'])

# 计算技术指标
# 1. 移动平均线
data['MA_5'] = data['close'].rolling(window=5).mean()
data['MA_10'] = data['close'].rolling(window=10).mean()
data['MA_20'] = data['close'].rolling(window=20).mean()

# 2. 布林带
data['std_20'] = data['close'].rolling(window=20).std()
data['upper_band'] = data['MA_20'] + 2 * data['std_20']
data['lower_band'] = data['MA_20'] - 2 * data['std_20']

# 3. RSI
delta = data['close'].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data['rsi'] = 100 - (100 / (1 + rs))

# 4. 收盘价与开盘价差的绝对值
data['open_close_diff'] = abs(data['close'] - data['open'])

# 5. 成交量加权平均价格（VWAP）
data['vwap'] = np.cumsum(data['volume'] * (data['high'] + data['low']) / 2) / np.cumsum(data['volume'])

# 6. MACD
exp12 = data['close'].ewm(span=12, adjust=False).mean()
exp26 = data['close'].ewm(span=26, adjust=False).mean()
data['macd'] = exp12 - exp26
data['macd_signal'] = data['macd'].ewm(span=9, adjust=False).mean()

# 7. ATR
data['tr'] = np.maximum((data['high'] - data['low']), np.maximum(abs(data['high'] - data['close'].shift(1)), abs(data['low'] - data['close'].shift(1))))
data['atr'] = data['tr'].rolling(window=14).mean()

# 8. Momentum
data['momentum'] = data['close'] - data['close'].shift(10)

# 9. OBV
data['obv'] = np.where(data['close'] > data['close'].shift(1), data['volume'], np.where(data['close'] < data['close'].shift(1), -data['volume'], 0)).cumsum()


                                                                                  
# 10. DI-
data['prev_high'] = data['high'].shift(1)
data['prev_low'] = data['low'].shift(1)
data['delta_high'] = data['prev_high'] - data['high']
data['delta_low'] = data['low'] - data['prev_low']
data['dm_minus'] = np.where((data['delta_high'] < 0) & (data['delta_low'] < 0) & (data['delta_low'] < data['delta_high']), data['delta_low'], 0)
data['dm_minus'] = np.where((data['delta_high'] < 0) & (data['delta_low'] < 0) & (data['delta_low'] >= data['delta_high']), 0, data['dm_minus'])
data['dm_minus'] = np.where(data['delta_high'] >= 0, 0, data['dm_minus'])

data['dm_minus_smooth'] = data['dm_minus'].rolling(window=14).sum()
data['tr_smooth'] = data['tr'].rolling(window=14).sum()

data['di_minus'] = 100 * (data['dm_minus_smooth'] / data['tr_smooth'])

# 清理临时列
data.drop(columns=['prev_high', 'prev_low', 'delta_high', 'delta_low', 'dm_minus', 'dm_minus_smooth', 'tr_smooth'], inplace=True)

# 输出结果
print(data.head())

# 将结果保存到 CSV 文件
data.to_csv("output.csv", index=False)

#移动平均线、布林带、相对强弱指数（RSI）、收盘价与开盘价差的绝对值、成交量加权平均价格（VWAP）、指数平滑异同平均（MACD）、平均真实波幅（ATR）、动量指标（Momentum）、震荡量指标（On Balance Volume，OBV）和负向差（DI-）
```
